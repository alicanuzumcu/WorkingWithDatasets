Finding **'MA7_P'** and **'LAG7_P'**

```python
today = 8 #datetime.day kullanÄ±p yazacaksÄ±n 8 yerine
tomorrow = today-1
df5 = merged_df[(merged_df['brand_id']==0)&(merged_df['product_id']==0)&(merged_df['store_id']==0)]
MA7_P = df5['quantity'][(tomorrow-7):tomorrow].sum()/7
MA7_P
LAG7_P = df5[df5['date'] == '2021-01-01']
LAG7_P = LAG7_P['quantity']
LAG7_P = LAG7_P[0]
message = f'MA7_P :{MA7_P}    LAG7_P :{LAG7_P}'
print(message)
```

Finding **'MA7_B'** and **'LAG7_B'**

```python
start_date = '2021-01-01'
end_date = '2021-01-07'

df6 = merged_df[(merged_df['brand_id']==0)&(merged_df['store_id']==0)]

filtered_df = df6[(df6['date'] >= start_date) & (df6['date'] <= end_date)]
MA7_B = filtered_df['quantity'].sum()/7
filtered_df2 = df6[(df6['date'] == start_date)]
LAG7_B = filtered_df2['quantity'].sum()
message = f'MA7_B :{MA7_B}    LAG7_B :{LAG7_B}'
print(message)
```

Finding **'MA7_S'** and **'LAG7_S'**

```python
start_date = '2021-01-01'
end_date = '2021-01-07'


df7 = merged_df[(merged_df['brand_id']==0)&(merged_df['store_id']==0)]
df8 = merged_df[(merged_df['brand_id']==1)&(merged_df['store_id']==0)]
df_sum = pd.concat([df7, df8], ignore_index=True)
sales_store_df = df_sum[(df_sum['date'] >= start_date) & (df_sum['date'] <= end_date)]
MA7_S = sales_store_df['quantity'].sum()/7
#MA7_S
sales_store_df2 = sales_store_df[(sales_store_df['date'] == start_date)]
LAG7_S = sales_store_df2['quantity'].sum()

message = f'MA7_S :{MA7_S}    LAG7_P :{LAG7_S}'
print(message)
```
